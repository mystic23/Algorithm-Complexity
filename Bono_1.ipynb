#Programar el algoritmo Bubble Sort y sacarle la complejidad

##1. Encontrando T(n)

$T(n) = \sum_{i=0}^{n-2}(\sum_{j=0}^{n-i-2}1))$

$T(n) = \sum_{i=0}^{n-2}(n-i-1)$

Hice varios despejes en el cuaderno, pero solo pondre como llegue

En secuencia $(n-1) + (n-2) + (n-3)+...+1 = \frac{n \cdot (n-1) }{ 2 } $

$T(n) = (n-1) \cdot \frac{n \cdot (n-1)}{2} = \frac{(n-1)^2 \cdot n}{2}$

Entonces tenemos que: $T(n) = O(n^2) $

```python
def bubbleSort(arr):
    n = len(arr)

    swapped = False

    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

        if not swapped:
            return
```

```python
import random
import time

list_sizes = [0,500,1000,3000,4000,5000]
sorted_times = []

for size in list_sizes:
    data = [random.randint(-1000, 1000) for _ in range(size)]

    start_time = time.time()
    bubbleSort(data)
    end_time = time.time()

    execution_time = end_time - start_time
    sorted_times.append(execution_time)

    # print(f"Tiempo de ejecución para lista de tamaño {size}: {execution_time:.6f} segundos")

print("Tiempos de ejecución:", sorted_times)
print("Tiempo de ejecución:", execution_time, "segundos")


```

```python
import matplotlib.pyplot as plt
plt.plot(list_sizes, sorted_times, marker='o')
plt.xlabel('Tamaño de la lista')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.title('Bubble sort')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time

def equation(n):
    return (n - 1)**2 * n / 2

list_sizes = [0, 500, 1000, 3000, 4000, 5000]

sorted_times = []
for size in list_sizes:
    data = [random.randint(-1000, 1000) for _ in range(size)]

    start_time = time.time()
    bubbleSort(data)
    end_time = time.time()

    execution_time = end_time - start_time
    sorted_times.append(execution_time)

n_values = np.array(list_sizes)
equation_values = equation(n_values)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(n_values, sorted_times, marker='o', color='blue')
plt.title('Bubble sort')
plt.xlabel('Tamaño de la lista')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(n_values, equation_values, marker='o', color='red')
plt.title('T(n) = (n-1)^2 * n / 2')
plt.xlabel('Valor de n')
plt.ylabel('Valor de la ecuación')
plt.grid(True)

plt.tight_layout()
plt.show()
```
